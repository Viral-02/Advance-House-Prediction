#### Categorical  Missing Feature

```python
categorical_features_nan = [feature for feature in df_test.columns if df_test[feature].dtypes == 'O' and df_test[feature].isnull().sum()>0]
```

```python
for feature in categorical_features_nan:
    print("Missing value for {}: {} %".format(feature,round(df_test[feature].isnull().mean()*100,2)))
```

```python
def categorical_missing_value(dataset,feature):
    feature_missing_mean = round(dataset[feature].isnull().mean()*100,2)
    if feature_missing_mean <= 10.0:
        mode_value = dataset[feature].mode()[0]
        dataset[feature].fillna(mode_value,inplace=True)
    elif feature_missing_mean > 10.0 and feature_missing_mean <= 51.0:
        dataset[feature].fillna("Missing",inplace=True)
    elif feature_missing_mean > 51.0:
        print(feature)
        dataset.drop(feature,axis=1,inplace=True)
```

```python
for feature in categorical_features_nan:
    categorical_missing_value(df_test,feature)
```

#### Numerical Missing Feature

```python
numerical_features = [feature for feature in df_test if df_test[feature].dtypes != 'O']
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
numerical_with_nan=[feature for feature in df_test.columns if df_test[feature].isnull().sum()>0 and df_test[feature].dtypes!='O']
discrete_feature = [feature for feature in numerical_features if len(df_test[feature].unique())<25 and feature not in year_feature+['Id']]
continuous_feature = [feature for feature in numerical_features if feature not in discrete_feature+year_feature+['Id']]
```

```python
sns.heatmap(df_test[numerical_features].isnull(),yticklabels=False)
```

```python
for feature in numerical_with_nan:
    if feature in continuous_feature:
        mean_value = df_test[feature].mean()
        df_test[feature].fillna(mean_value,inplace=True)
        plt.axvline(x= df_test[feature].mean(),ls='--',color='yellow')
        plt.axvline(x= df_test[feature].median(),ls='--',color='Red')
        df_test[feature].hist(bins=25)
        plt.xlabel(feature)
        plt.ylabel("Count")
        plt.title(feature)
        plt.show()
    else:
        mode_value = df_test[feature].mode()
        df_test[feature].fillna(mode_value,inplace=True)
```

```python
for feature in year_feature:
    if feature != 'YrSold':
        df_test[feature] = df_test['YrSold'] - df_test[feature]
```

```python
## As Yearbuilt and garageYrBlt has same impact and highly correlated we can drop this particular feature
df_test.drop('GarageYrBlt',axis=1,inplace=True)
```
